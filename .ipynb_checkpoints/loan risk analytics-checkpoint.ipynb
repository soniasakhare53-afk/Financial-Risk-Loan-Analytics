```python
import mysql.connector
import pandas as pd

# MySQL connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",              # change if different
    password="sakharesonia2008", # replace
    database="loan_risk_analytics"
)

print("Connected to MySQL successfully!")
```

```python
query = """
SELECT 
    c.customer_id,
    c.name,
    c.annual_income,
    cs.credit_score,
    COUNT(CASE WHEN r.payment_status = 'MISSED' THEN 1 END) AS missed_emis,
    SUM(l.loan_amount) AS total_loan_amount,
    CASE WHEN d.loan_id IS NOT NULL THEN 1 ELSE 0 END AS defaulted
FROM customer c
JOIN loans l ON c.customer_id = l.customer_id
LEFT JOIN repayments r ON l.loan_id = r.loan_id
LEFT JOIN credit_scores cs ON c.customer_id = cs.customer_id
LEFT JOIN defaults d ON l.loan_id = d.loan_id
GROUP BY c.customer_id, c.name, c.annual_income, cs.credit_score, defaulted
"""

df = pd.read_sql(query, conn)
df
```

## Calculate Risk Score

```python
def calculate_risk(row):
    score = 0
    
    # Missed EMI risk
    if row['missed_emis'] >= 2:
        score += 40
    elif row['missed_emis'] == 1:
        score += 20
    
    # Credit score risk
    if row['credit_score'] < 650:
        score += 30
    elif row['credit_score'] < 700:
        score += 15
    
    # Loan to income risk
    loan_income_ratio = row['total_loan_amount'] / row['annual_income']
    if loan_income_ratio > 0.6:
        score += 25
    elif loan_income_ratio > 0.4:
        score += 15
    
    # Past default risk
    if row['defaulted'] == 1:
        score += 50
        
    return score

df['risk_score'] = df.apply(calculate_risk, axis=1)
df
```

## Convert Risk Score â†’ Risk Category

```python
def risk_category(score):
    if score >= 70:
        return 'HIGH'
    elif score >= 40:
        return 'MEDIUM'
    else:
        return 'LOW'

df['risk_category'] = df['risk_score'].apply(risk_category)
df[['customer_id', 'name', 'risk_score', 'risk_category']]
```

## Risk Category Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='risk_category', data=df)
plt.title("Customer Risk Category Distribution")
plt.show()
```

## Risk Score vs Credit Score

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='credit_score',
    y='risk_score',
    hue='risk_category',
    data=df
)
plt.title("Credit Score vs Risk Score")
plt.show()
```

## Loan Burden vs Risk

```python
df['loan_income_ratio'] = df['total_loan_amount'] / df['annual_income']

plt.figure(figsize=(6,4))
sns.boxplot(x='risk_category', y='loan_income_ratio', data=df)
plt.title("Loan-to-Income Ratio by Risk Category")
plt.show()
```
